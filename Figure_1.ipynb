```python
import numpy as np
import matplotlib.pyplot as plt
```

### Load data

```python
Dict = dict(np.load('./Data/Fig_1_data.npy', allow_pickle=True).item())
Frequency = Dict['Frequency']
Population = Dict['Population']
```

### Plot

```python
fig, ax = plt.subplots(figsize=(5,2), dpi=150)
ax.plot(Frequency, Population)
ax.set_xlabel('Drive frequency (GHz)')
ax.set_ylabel('$|2\\rangle$ Population')
```
